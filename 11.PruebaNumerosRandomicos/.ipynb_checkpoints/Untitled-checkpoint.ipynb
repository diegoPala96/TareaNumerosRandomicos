## Metodo cuadrado medios


```python
import time
import psutil
import numpy as np
import matplotlib.pyplot as plt
print("digitos")
digitos=int(input())

frac=10**digitos
my_list = []
list2=[]
interacioninterna=0
for i in range(100):
    print("interacion numero:",i+1)
    xn=psutil.virtual_memory()[3] 
    #xn=time.time() #generar una semilla diferente por cada iteracion
    print("semilla:",xn)
    while(True):
        interacioninterna=interacioninterna+1
        #if(a==3):
         #   a=0
          #  break
        xn2=xn*xn
        cadena=str(xn2).zfill(8)
        longitud=int(len(str(cadena)))
        medio=int(longitud/2)
        ui=int(cadena[(medio-2):medio]+cadena[medio:(medio+2)])
        rn=ui/frac
        #print(interacioninterna,"\t \t",xn,"\t",cadena,"\t", longitud,"\t",str(ui).zfill(4),"\t",rn,ui in my_list)
        xn=ui #
        #print((xn in my_list) is True)
        if ((xn in my_list) is True):
            list2.append(interacioninterna)
            #a=1
            #print('si')
            break
        my_list.append(xn)
    #a=0
    my_list = []
    interacioninterna=0
   
    #i=i+1
       
    
    
```

```python
def contarElementos(lista):
    return{i:lista.count(i) for i in lista}
list2.sort()
listnueva=contarElementos(list2)
print(listnueva)
```

```python
d = listnueva
X = np.arange(len(d))
plt.bar(X, d.values(), align='center', width=0.5)
plt.xticks(X, d.keys())

ymax = max(d.values()) + 1
plt.ylim(0, ymax)
plt.title("metodo cuadrado medios")
plt.show()
```
